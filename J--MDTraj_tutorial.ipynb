<h3>Computing Root Mean Square Deviation (RMSD)</h3>

One way we might want to analyze our trajectory is by performing a Root Mean Squared Deviation (rmsd) calculation of the entire trajectory.  Below we compute the differences between each frame of the trajectory and the first frame (frame 0). 

```python
# The rmsd method has the following format. 
# mdtraj.rmsd(target, reference, frame=0, atom_indices=None)

# (you can always check the signature of a function with IPython by
# putting a question mark after the function name and hitting shift-enter)

combined_rmsd_calculated = mdtraj.rmsd(combined_traj, combined_traj, frame=0)
```

We could examine the RMSD data, but a visual representation may enable us to distinguish patterns more readily.  A handy library called matplotlib provides easy-to-use graphing functions.
We imported the library earlier (import matplotlib.pyplot as plt) and told ipython that we would like to see the graphics in the browser (%pylab inline).  These commands only need to be run once within a given notebook.

```python
#Let's plot the raw data for the rmsd.
plt.plot(combined_rmsd_calculated, color='blue')
plt.xlabel('Time')
plt.ylabel('RMSD(nm)')
plt.title('RMSD')
plt.show()
```

This doesn't seem too useful since the RMSD doesn't seem to be changing that much in trajectory 1 (time 0-399) or trajectory 2 (time 400-799). This can be seen from the two state picture as well since only the turn near the C-terminus is actually exhibiting any motion. Let's zoom in and only compute the RMSD for the loop.  Using MDTraj, we previously determined that indices 113 to 202 correspond to the loop.

#   <h1>Plotting Dihedrals</h1>

Now let's play with the data a little more. After some initial viewings of the trajectory, you realize that this loop motion can be quantified using the 4th phi (5th if you start counting from 1;  4th if you start counting from 0) dihedral of the system. Lucky for you, MDTraj can compute phi,psi,chi1,chi2 or pretty much any arbitrary dihedral. 

```python
# compute the indices and the dihedrals 
indices_list, phi_dihed =  mdtraj.compute_phi(combined_traj)
#the mdtraj.compute_phi function returns two things 
# 1) the indices used to compute the phi dihedrals(called indices_list here)
# 2) an m*n matrix (called phi_dihed) which has the m-rows corresponding to the 'm' time points(800 here), and n-columns(corresponding to the number of phi dihedrals in the system)

# we can plot the dihedrals using the same method as above 
dihedral_wanted = 4
##note the change in the plotting command, we are telling python that we would like to plot all the "m" data points using the special ":" symbol but that we would only like to plot 
### the values from the 4th column(the 4th dihedral) and that we would like it in red. 
plt.plot(phi_dihed[:, dihedral_wanted], "red")

plt.ylabel('radians')
plt.xlabel('time')
plt.show()
```
